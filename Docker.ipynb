### docker container 생성 실행
- 컨테이너 실행은 터미널이나 ps 이용하지 않고 vscode나 docker dsek에서 실행
- 생성 명령 : dcker create 옵션 컨테이너명 이미지명
- 옵션
    - -it : 컨테이너 실행하고 샐행을 종료하지 말고 유지시킬 것
    - --name : 생성되는 컨테이너의 이름을 부여
- 코드 : docker create -it --name jupyter_ubuntu_20.04 ubuntu:20.24

### docker start 컨테이너 이름
- docker start jupyter_ubuntu_20.04




### 주피터 실행

### 리눅스 필수 명령어
- cat /etc/shells : 현재 시스템에서 사용할 수 있는 shell 프로그램 확인
- echo \$SHELL : 현재 기본값으로 설정되어 사용중이 shell(터미널) 프로그램 확인
    - echo : 변수내의 값을 던져줄 것
    - $로 시작하는 환경변수 : 시스템 어느 위치에서나 실행이 가능하도록 시스템 환경에 등록해 놓는 변수
- 계정별 설정 파일
    - 리눅스(ubuntu)의 설정파일 : bashrc
    - /etc/barshrc : 전역 설정 파일(시스템(전체계정)에 영향을 미치는 설정을 저장)
    - 없으면 생성해서 사용하면 됨
    - 계정별 설정 파일
    - 계정홈 /.bashrc
    - 계정홈 : root는 /root 디렉토리, 일반계정은 /home/ 
    - 파일명 앞에 .이 있으면 숨김파일이라는 의미

### file과 관련된 일반 명령어
- 빈 파일 생성
    - touch 파일명 : 빈파일 생성
    - cat 파일명 : 파일 내용을 출력
    
- 파일 복사
    - cp 원본파일경로및파일명 복사할경로및파일명
    - cp /etc/hosts hosts
    
- 현재 디렉토리에 /etc/passwd 파일을 password라는 파일명으로 복사하시오.
    - cp /ect/passwd password
    - 복사는 권한과 관련있을 수 있음(읽기권한과 관련있음)
    
- 파일이동
    - mv 원본파일경로및파일명 복사할경로및파일명

- 아래 이동코드는 경로가 변경되지 않는 코디이므로 팔명만 변경됨
