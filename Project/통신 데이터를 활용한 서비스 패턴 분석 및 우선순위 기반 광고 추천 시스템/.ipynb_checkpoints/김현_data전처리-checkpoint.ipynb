# 2023년1월~9월까지의 9개 데이터 합치기

```python
import pandas as pd
import numpy as np

# 파일 경로
file_path = "./2023.{}월_29개 통신정보.xlsx"

# 1월부터 9월까지의 데이터를 읽어오는 리스트를 만들기
dataframes = [pd.read_excel(file_path.format(month)) for month in range(1, 10)]

# 각 데이터프레임을 하나로 합치기
combined_df = pd.concat(dataframes)

# 숫자 데이터만 포함하는 열을 선택
numeric_df = combined_df.select_dtypes(include=[np.number])

# 동일한 위치의 셀들을 묶어 평균을 계산
average_df = numeric_df.groupby(level=0).mean()

# 결과를 새로운 엑셀 파일로 저장
average_df.to_excel('./average.xlsx')
```

# 행정동에서 자치구로 통합하고 성별 나이대별로 세분화 작업

```python
import pandas as pd
import numpy as np

# 숫자형 컬럼만 선택
numeric_df = df.select_dtypes(include=[np.number])

# 자치구, 연령대, 성별로 그룹화
grouped_df = numeric_df.groupby([df['자치구'], df['연령대'], df['성별']])

# 각 그룹의 평균 계산
average_df = grouped_df.mean()

# 새로운 엑셀 파일로 저장
average_df.to_excel('average_grouped.xlsx')

print("'average_grouped.xlsx' 파일에 데이터가 저장되었습니다.")
```

# 평균연령대구하기(자치구별 평균연령대 + 자치구 성별 평균연령대)

```python
import pandas as pd

# 데이터 불러오기
df = pd.read_excel('average_grouped.xlsx')

# 성별에 따른 자치구별 가중 평균 연령 계산
df_sex_age = df.groupby(['자치구', '성별']).apply(lambda x: (x['연령대'] * x['총인구수']).sum() / x['총인구수'].sum()).reset_index()
df_sex_age.columns = ['자치구', '성별', '성별평균연령대']

# 자치구별 총 가중 평균 연령 계산
df_total_age = df.groupby('자치구').apply(lambda x: (x['연령대'] * x['총인구수']).sum() / x['총인구수'].sum()).reset_index()
df_total_age.columns = ['자치구', '총평균연령대']

# 두 DataFrame을 자치구 기준으로 병합
result_df = pd.merge(df_sex_age, df_total_age, on='자치구')

# 기존 DataFrame에 결과 DataFrame 병합
df_final = pd.merge(df, result_df,  how='left', left_on=['자치구','성별'], right_on = ['자치구','성별'])

# 결과 저장
df_final.to_excel('average_age_result.xlsx', index=False)

# 완료 메시지 출력
print("성별 평균 연령대 및 총 평균 연령대 계산 및 저장이 완료되었습니다.")
```

# 그래프 20~75 전체 자치구별 총 인구수 비교

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# 한글 폰트 설정
font_path = 'C:/Windows/Fonts/malgun.ttf'  # 폰트 경로를 알맞게 설정해주세요.
font_prop = fm.FontProperties(fname=font_path).get_name()
plt.rc('font', family=font_prop)

# 엑셀 파일 읽어오기
df = pd.read_excel('average_age_result.xlsx')

# 자치구, 연령대별로 그룹화하여 총인구수 합산
df_grouped = df.groupby(['자치구', '연령대']).sum().reset_index()

# 그래프 그리는 함수 정의
def plot_graph(age, color):
    df_age = df_grouped[df_grouped['연령대'] == age]
    plt.figure(figsize=(12, 6))
    plt.bar(df_age['자치구'], df_age['총인구수'], label=f'{age}대', color=color)
    plt.xlabel('자치구')
    plt.ylabel('총인구수')
    plt.title(f'{age}대 자치구별 총인구수 비교')
    plt.xticks(rotation=45)
    plt.legend()
    plt.show()

# 연령대별로 그래프 생성
colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k']  # 색상 리스트
age_range = list(range(20, 76, 5))
for i, age in enumerate(age_range):
    plot_graph(age, colors[i % len(colors)])
```

# 그래프 하나하나씩 표기하고싶을때 ex) 20대만

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# 한글 폰트 설정
font_path = 'C:/Windows/Fonts/malgun.ttf'  # 폰트 경로를 알맞게 설정해주세요.
font_prop = fm.FontProperties(fname=font_path).get_name()
plt.rc('font', family=font_prop)

# 엑셀 파일 읽어오기
df = pd.read_excel('average_age_result.xlsx')

# 자치구, 연령대별로 그룹화하여 총인구수 합산
df_grouped = df.groupby(['자치구', '연령대']).sum().reset_index()

# 연령대 필터링: 20대만
df_age_20s = df_grouped[df_grouped['연령대'] == 20]

# 바플롯 생성
plt.figure(figsize=(12, 6))
plt.bar(df_age_20s['자치구'], df_age_20s['총인구수'], label='20대')
plt.xlabel('자치구')
plt.ylabel('총인구수')
plt.title('20대 자치구별 총인구수 비교')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```
