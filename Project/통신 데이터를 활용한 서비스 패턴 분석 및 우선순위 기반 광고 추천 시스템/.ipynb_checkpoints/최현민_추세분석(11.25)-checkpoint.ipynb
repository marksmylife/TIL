```python
import pandas as pd

# 파일 경로 포맷 문자열
file_path_format = "./data/2023.{}월_29개 통신정보.xlsx", usecols=

# 추출할 열 이름
columns_to_extract = ['자치구', '성별', '연령대', '총인구수',
                      '게임 서비스 사용일수', '금융 서비스 사용일수', '배달 서비스 사용일수', '쇼핑 서비스 사용일수',
                      '동영상/방송 서비스 사용일수', '넷플릭스 사용일수', '유튜브 사용일수']

# 연령대를 10년 단위로 그룹화하는 함수
def merge_age(age):
    return int((age // 10) * 10)

age_group_means = {}

# 연령대와 성별로 그룹화한 후 평균을 계산하는 코드
for month in range(1, 10):
    file_path = file_path_format.format(month)
    month_str = f"{month}월"
    
    df = pd.read_excel(file_path, usecols=columns_to_extract)
    df['성별'] = df['성별'].map({1: 'M', 2: 'F'})
    df['연령대'] = df['연령대'].apply(merge_age)

    # '자치구' 열을 삭제
    df = df.drop(['자치구'], axis=1)
    
    # 연령대와 성별로 그룹화한 후 평균을 계산
    age_group_mean = df.groupby(['연령대', '성별']).mean().reset_index()

    # 결과를 딕셔너리에 저장
    age_group_means[f"{month_str} mean"] = age_group_mean

#     # 연령대별 평균 데이터를 엑셀 파일로 저장
#     output_file_path = f"./2023.{month}월_연령대별_평균.xlsx"
#     age_group_mean.to_excel(output_file_path, index=False)


# 예시 출력: 1월 데이터의 연령대별 평균
age_group_means['1월 mean'].head(10)
```

```python
# 연령대별 평균 데이터프레임을 월별로 정렬
months_ordered = [f"{m}월 mean" for m in range(1, 10)]
dfs_ordered = [age_group_means[month] for month in months_ordered]

# 증감수치와 증감률을 계산하는 함수
def calculate_growth(previous, current):
    growth = current - previous
    growth = np.round(growth, 3)
    # numpy의 where 함수를 사용하여 벡터화된 조건부 연산 수행
    growth_rate = np.where(previous != 0, (growth / previous) * 100, float('inf'))
    return growth, growth_rate

# 모든 열에 대해 증감수치와 증감률을 계산하고 결과를 저장할 데이터프레임을 만듬
growth_df_list = []

# 첫 번째 월은 비교 대상이 없으므로 증감률 계산에서 제외
for i in range(1, len(dfs_ordered)):
    previous_df = dfs_ordered[i - 1]
    current_df = dfs_ordered[i]
    growth_df = pd.DataFrame()
    
    # 연령대와 성별은 그대로 유지
    growth_df['연령대'] = current_df['연령대']
    growth_df['성별'] = current_df['성별']
    
    # 서비스 사용일수 열에 대해 증감수치와 증감률 계산
    for col in current_df.columns[2:]:  # '연령대'와 '성별'을 제외한 나머지 열에 대해 계산
        growth, growth_rate = calculate_growth(previous_df[col], current_df[col])
        growth_df[col + ' 증감수치(증감률)'] = growth.astype(str) + "(" + growth_rate.round(2).astype(str) + "%)"
    
    # 해당 월을 표시하는 열 추가
    growth_df['월'] = f"{i+1}월"
    growth_df_list.append(growth_df)

# 모든 증감 데이터프레임을 하나로 합침
all_growth_df = pd.concat(growth_df_list, ignore_index=True)
```

```python
# 결과 확인
print(all_growth_df.head())
all_growth_df.to_excel('./증감률.xlsx')
```

```python
import pandas as pd
import numpy as np

# 연령대별 평균 데이터프레임을 월별로 정렬
months_ordered = [f"{m}월 mean" for m in range(1, 10)]
dfs_ordered = [age_group_means[month] for month in months_ordered]

# 결과를 저장할 데이터프레임을 초기화합니다.
final_growth_df = dfs_ordered[0][['연령대', '성별']].copy()

for i in range(1, len(dfs_ordered)):
    previous_df = dfs_ordered[i - 1]
    current_df = dfs_ordered[i]
    # 증감 계산
    growth = current_df.iloc[:, 2:] - previous_df.iloc[:, 2:]
    # 증감률 계산
    growth_rate = growth / previous_df.iloc[:, 2:].replace(0, np.nan) * 100
    
    # 소수점 세 번째 자리에서 반올림
    growth = growth.round(3)
    growth_rate = growth_rate.round(2)

    # 증감수치(증감률) 형식으로 문자열 결합
    growth_formatted = growth.astype(str) + "(" + growth_rate.astype(str) + "%)"
    
    # 월별 증감수치를 데이터프레임에 추가
    for col in growth_formatted.columns:
        final_growth_df[f'{i}월~{i+1}월_{col}'] = growth_formatted[col].values

# 결과 출력
print(final_growth_df)

final_growth_df.to_excel('./final_growth_df.xlsx')
```
