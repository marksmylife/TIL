```python
data= pd.read_excel('average.xlsx')
data.head()
```

```python
df = pd.DataFrame(data)
data1 = df[['자치구','성별','연령대','총인구수']]
data1
```

```python
data=data1.replace(1,'f')
data1=data.replace(2,'m')
data1
```

- 평균연령대
 - 남/여/총합


```python
#총합
av_age=(data1['연령대'] * data1['총인구수']).sum()/data1['총인구수'].sum()
av_age
```

```python
f=data1[(data1.성별=='f')]
f.head()
f.shape
```

```python
#전체 남성
f_age=(f['연령대'] * f['총인구수']).sum()/f['총인구수'].sum()
f_age
```

```python
m=data1[(data1.성별=='m')]
m.head()
m.shape
```

```python
#전체 여성
m_age=(m['연령대'] * m['총인구수']).sum()/m['총인구수'].sum()
m_age
```

```python
#총합 데이터
age= pd.Series()
age['남자평균연령'] = f_age
age['여자평균연령'] = m_age
age['총합평균연령'] = av_age
```

```python
#평균연령대 bar그래프
import matplotlib.pyplot as plt

plt.bar(age.index, color='skyblue', label='남자')
plt.bar(age['여자평균연령'], color='orange', label='여자',)

# 그래프에 제목과 축 레이블 추가
plt.title('평균연령대bar그래프')
plt.xlabel('성별')
plt.ylabel('평균연령대')

# 그래프 출력
plt.show()
```
