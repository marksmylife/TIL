#### Actors and Use Cases
* 액터(Actor)
    * 액터는 시스템 외부에 있는 누군가(또는 무엇인가)로, 시스템과 상호작용합니다.
    * "누군가"는 시스템을 사용하는 사람의 역할을 의미합니다. 이는 고객, 관리자, 다른 시스템, 하드웨어 장치 등 다양한 형태일 수 있습니다.
    * 액터는 시스템의 외부에 위치하며, 시스템의 서비스나 기능을 사용하려는 주체입니다.
    * 예를 들어, 온라인 쇼핑몰 시스템에서 고객, 배송 서비스, 결제 시스템 등이 액터가 될 수 있습니다.

* 유스케이스(Use Case)
    * 유스케이스는 액터가 시스템을 사용하여 수행하고자 하는 작업이나 과정을 설명합니다.
    * 이는 액터가 시스템에 접근하여 달성하고자 하는 구체적인 목표나 요구사항을 반영합니다.
    * 유스케이스는 일반적으로 시나리오 또는 사용 사례 형태로 작성되며, 시스템이 제공해야 할 서비스 또는 기능을 명시합니다.
    * 예를 들어, 은행 시스템에서 "계좌 조회", "송금", "예금" 등이 유스케이스가 될 수 있습니다.

#### Communicates-Association
* usecase 다이어그램 그릴 시 서로 순서가 있음 안됨.
* usecase는 단독적으로 수행되어야 함, 즉 순서가 있으면 안된다.
* 순서는 actor가 수행할 일임.
* 선 또는 화살표 사용
    * 액터와 유스케이스 사이의 상호작용은 선이나 화살표로 표현.
    * 이는 두 요소 간의 연결을 나타내며, 시스템과 사용자(또는 다른 액터) 사이의 상호작용을 시각화합니다.

* 화살표의 의미
    * 화살표는 통신을 시작하는 주체(function call)를 나타냅니다.(일방향)
        * 주로 actor가 '사람'일 경우
    * 화살표가 없고 선만 있으면 **양방향**임
    * 즉, 화살표가 액터에서 유스케이스를 가리키고 있다면, 그 액터가 작용을 시작한다는 것을 의미
    * 이는 액터가 유스케이스에 해당하는 시스템의 기능이나 서비스를 요구하는 상황을 나타냅니다.

* 예시
    * 예를 들어, "고객" 액터가 "상품 주문" 유스케이스와 연결된 경우, 화살표가 고객에서 상품 주문 유스케이스로 향하고 있다면, 고객이 주문 프로세스를 시작한다는 것을 나타냅니다.

* 프리컨디션
    * 유스케이스를 시작하기 전에 충족되어야 하는 조건을 말합니다. 
    * 예를 들어, "계좌에서 돈을 인출하기 전에 사용자가 로그인해야 한다"는 로그인 유스케이스가 인출 유스케이스보다 먼저 수행되어야 함을 나타냅니다.

* 포스트컨디션
    * 유스케이스가 완료된 후의 시스템 상태를 설명합니다.
    * 이는 후속 유스케이스의 프리컨디션이 될 수 있습니다.
    
* 하지만 프리, 포스트컨디션을 다이어그램에 명시하는 것은 아니다!, 다이어그램에 모든 것을 표시하면 안된다!

#### Association Navigability between Actors and Use Cases
*  Association Navigability = interaction invocatation의 방향
* 사람의 역할을 대변하는 Actor는 대부분의 경우(Active Actor)
    * Actor > Use Case
* 시스템 > User에게 Alarm, Request Message, News Feed등을 보내는 경우, 사람의 역할을 대변하는 Actor가 (Passive Actor)
    * Use Case > Actor : 증권 알람이나 결재 알람 등
* 시스템 대변 Actor는 어떻게 하는거지???

#### Definition of Use Cases
* **A use case defines a sequence of actions performed by a system that yields an observable result of value to an actor.**
* 각 유즈 케이스는 특정한 사용자 상호작용을 나타내며, 시스템이 제공해야 하는 결과를 명확하게 합니다.
* 사람의 입력이 아니라 시스템의 리액션을 찾야아함.
    * 예) 쇼핑몰에서 고객이 구입하는 것이 아니라, 시스템에서 재고, 배송예약 등의 리액션을 찾아야한다.
* 유스케이스 > 스페시픽(유스케이스의 결과물이 적혀있어야 함)
* 유스케이스를 검증하는 모든 것은 유스케이스의 definition에 있다.
